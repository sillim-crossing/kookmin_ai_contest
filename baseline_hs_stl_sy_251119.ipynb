## Step 1.1: STL분해 및 residual only 패턴

````python
from statsmodels.tsa.seasonal import STL

# ==============================
# STL residual pivot 만들기
# ==============================

def extract_residual(series, period=12):
    """
    pivot.loc[item_id] 1개 시계열을 받아 residual만 반환
    """
    s = pd.Series(series)

    # 0 → NaN 처리 후 보간
    s = s.replace(0, np.nan).interpolate().fillna(method="bfill").fillna(method="ffill")

    stl = STL(s, period=period, robust=True).fit()
    return stl.resid

# residual pivot 생성
residual_pivot = pd.DataFrame(
    {
        item: extract_residual(pivot.loc[item])
        for item in pivot.index
    }
).T

print("원본 pivot:", pivot.shape)
print("Residual pivot:", residual_pivot.shape)
````

## Step 3: 통계적 유의성 검증

````python
from scipy.stats import pearsonr

def safe_corr_with_pvalue(x, y):
    if np.std(x) == 0 or np.std(y) == 0:
        return 0.0, 1.0
    corr, p_value = pearsonr(x, y)
    return float(corr), float(p_value) # 피어슨 상관계수, p-value
````

## Step 4: Granger Causality 검증

- A의 과거 정보가 B의 미래를 예측하는 데 도움이 되는가?"**를 통계적으로 검증하는 방법



```
# B의 과거 + A의 과거 사용
B(3월) = f(B의 1월, B의 2월, A의 1월, A의 2월)
B(4월) = f(B의 2월, B의 3월, A의 2월, A의 3월)
...

예측 오차 = 2  ← 더 정확!
```

**결론**
```
A의 정보를 추가했더니 예측이 더 정확해짐!
→ "A가 B를 Granger-cause 한다" ✅
```

---

🔬 통계적 검증

**가설 검정**
```
H0 (귀무가설): A의 과거 정보는 B 예측에 도움 안 됨
H1 (대립가설): A의 과거 정보는 B 예측에 도움 됨

→ F-test로 검증
→ p-value < 0.05 이면 H0 기각 → A가 B를 cause함!
```



````python
from statsmodels.tsa.stattools import grangercausalitytests

def granger_test(x, y, max_lag=6):
    """
    x가 y를 Granger-cause 하는지 검증
    """
    try:
        data = pd.DataFrame({'x': x, 'y': y})
        test_result = grangercausalitytests(data[['y', 'x']],
                                            maxlag=max_lag,
                                            verbose=False)
        # 각 lag의 p-value 중 최소값 반환
        p_values = [test_result[lag][0]['ssr_ftest'][1]
                   for lag in range(1, max_lag+1)]
        return min(p_values)
    except:
        return 1.0
````

## Step 5: 공행성 강도 점수 계산

````python
def calculate_score_advanced(corr, p_value, granger_p, lag):
    """
    정교한 공행성 점수 계산 (상관계수 비중 대폭 증가)
    """

    # 1. 상관계수 점수 (0~85점) - 비선형 스케일링, 비중 대폭 증가
    # 강한 상관관계에 더 높은 보상
    if abs(corr) >= 0.8:
        corr_score = 85
    elif abs(corr) >= 0.7:
        corr_score = 75
    elif abs(corr) >= 0.6:
        corr_score = 62
    elif abs(corr) >= 0.5:
        corr_score = 48
    elif abs(corr) >= 0.4:
        corr_score = 35
    else:
        corr_score = abs(corr) * 85  # 0.4 미만은 선형

    # 2. 통계적 유의성 점수 (0~12점) - 비중 축소
    if p_value < 0.001:
        sig_score = 12  # 매우 강한 유의성
    elif p_value < 0.01:
        sig_score = 9
    elif p_value < 0.05:
        sig_score = 5
    else:
        sig_score = 2  # 약한 유의성

    # 3. Lag 점수 (0~3점) - 비중 축소
    # 최적 lag는 2-3개월로 가정
    if lag == 2 or lag == 3:
        lag_score = 3  # 최적
    elif lag == 1 or lag == 4:
        lag_score = 2  # 양호
    elif lag == 5:
        lag_score = 1  # 보통
    else:  # lag >= 6
        lag_score = 0  # 너무 김

    total_score = corr_score + sig_score + lag_score

    return total_score
````

## Step 6: 각 Leading별 Following 탐지

````python
from tqdm import tqdm

def find_comovement_pairs(
    pivot,
    max_lag=6,
    min_nonzero=12,
    corr_threshold=0.4,
    score_threshold=40
):
    items = pivot.index.to_list()
    months = pivot.columns.to_list()
    n_months = len(months)
    results = []

    for i, leader in tqdm(enumerate(items), total=len(items)):
        x = pivot.loc[leader].values.astype(float)
        if np.count_nonzero(x) < min_nonzero:
            continue

        candidates = []

        for follower in items:
            if follower == leader:
                continue

            y = pivot.loc[follower].values.astype(float)
            if np.count_nonzero(y) < min_nonzero:
                continue

            best_lag = None
            best_corr = 0.0
            best_p_value = 1.0

            # lag 탐색
            for lag in range(1, max_lag + 1):
                if n_months <= lag:
                    continue

                corr, p_value = safe_corr_with_pvalue(x[:-lag], y[lag:])

                if abs(corr) > abs(best_corr):
                    best_corr = corr
                    best_lag = lag
                    best_p_value = p_value

            # 기본 임계값 통과 시
            if best_lag is not None and abs(best_corr) >= corr_threshold:
                # Granger test
                granger_p = granger_test(x, y, max_lag=best_lag)

                # 종합 점수 계산
                score = calculate_score_advanced(
                    best_corr, best_p_value, granger_p, best_lag
                )

                # 점수 임계값 통과 시 후보에 추가
                if score >= score_threshold:
                    candidates.append({
                        "following_item_id": follower,
                        "best_lag": best_lag,
                        "max_corr": best_corr,
                        "p_value": best_p_value,
                        "granger_p_value": granger_p,
                        "comovement_score": score,
                    })

        # 알파벳 순으로 정렬
        candidates.sort(key=lambda x: x['following_item_id'])

        for candidate in candidates:
            results.append({
                "leading_item_id": leader,
                "following_item_id": candidate["following_item_id"],
                "best_lag": candidate["best_lag"],
                "max_corr": candidate["max_corr"]
            })

    pairs = pd.DataFrame(results)
    return pairs

# 실행
pairs = find_comovement_pairs(pivot,9,12,0.4,35)
print("탐색된 공행성쌍 수:", len(pairs))
pairs.head(20)

# CSV로 저장
date_str = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
pairs_file_name = f"pairs_sy_{date_str}.csv"
pairs.to_csv(pairs_file_name, index=False, encoding='utf-8-sig')
print(f"\n✅ 파일 저장 완료: {pairs_file_name}")

# Google Colab에서 다운로드
from google.colab import files
files.download(pairs_file_name)
````

# Step 7: DTW등 각 물품 주요 특징 피처 생성

````python
# ============================================================
# 7. DTW / volatility_corr / Granger 등 메타피처 생성
# ============================================================

print("\n===== 메타피처 생성 시작 =====")

PairsMeta_rows = []

for idx, row in tqdm(pairs.iterrows(), total=len(pairs)):
    leader = row["leading_item_id"]
    follower = row["following_item_id"]
    lag = int(row["best_lag"])
    corr_raw = float(row["max_corr"])

    # --------------------------
    # 1. 시계열 준비
    # --------------------------
    ts_L_raw = pivot.loc[leader].fillna(0).values
    ts_F_raw = pivot.loc[follower].fillna(0).values

    ts_L_res = residual_pivot.loc[leader]
    ts_F_res = residual_pivot.loc[follower]

    # --------------------------
    # 2. DTW (lag 반영)
    # --------------------------
    dtw_dist = np.nan
    if lag > 0 and lag < len(ts_L_raw):
        try:
            dtw_dist = dtw(ts_L_raw[:-lag], ts_F_raw[lag:])
        except:
            pass

    # --------------------------
    # 3. Granger p-value
    # --------------------------
    granger_p = np.nan
    try:
        df_g = pd.DataFrame({"F": ts_F_raw, "L": ts_L_raw})
        if lag > 0 and len(df_g) > lag + 5:
            result = grangercausalitytests(df_g[["F", "L"]], maxlag=[lag], verbose=False)
            granger_p = result[lag][0]["ssr_ftest"][1]
    except:
        pass

    # --------------------------
    # 4. residual volatility corr
    # --------------------------
    vol_corr = np.nan
    try:
        L_vol = ts_L_res.rolling(window=3).std().fillna(0)
        F_vol = ts_F_res.rolling(window=3).std().fillna(0)
        vol_corr = L_vol.corr(F_vol)
    except:
        pass

    # --------------------------
    # 5. Corr scaling
    # --------------------------
    corr_scaled = np.log1p(abs(corr_raw)) * np.sign(corr_raw)

    # --------------------------
    # 하나의 메타 row 저장
    # --------------------------
    PairsMeta_rows.append({
        "leading_item_id": leader,
        "following_item_id": follower,

        # 원본 pair meta
        "f_corr_raw": corr_raw,
        "f_corr_scaled": corr_scaled,
        "f_lag": lag,
        "f_is_common": row["f_is_common"],

        # 새로 생성된 메타피처
        "f_dtw_distance_raw": dtw_dist,
        "f_granger_pvalue": granger_p,
        "f_volatility_corr_resid": vol_corr,
    })


# 메타피처 테이블 완성
date_str = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
pairs_meta = pd.DataFrame(PairsMeta_rows)
print("pairs_meta 생성 완료 →", pairs_meta.shape)

# 저장(optional)
pairs_meta.to_csv(f"hs_pairs_meta_{date_str}.csv", index=False, encoding="utf-8-sig")
````

````python
# 시계열 피처를 pairs_meta에 추가하는 함수 (핵심 5개만)
def add_time_series_features(pairs_meta, pivot):
    """
    pairs_meta 데이터프레임에 핵심 시계열 피처 5개만 추가
    """
    months = pivot.columns.tolist()  # 월 컬럼 리스트
    features_list = []

    for idx, row in pairs_meta.iterrows():
        leader = row['leading_item_id']
        follower = row['following_item_id']
        lag = int(row['f_lag'])
        t = len(months) - 1  # 마지막 시점
        ref_idx = max(0, t - lag)

        # Leader 데이터
        L_raw = pivot.loc[leader, months[ref_idx]]
        L_last = pivot.loc[leader, months[t]]
        L_prev1 = pivot.loc[leader, months[t-1]] if t >= 1 else L_last

        # Follower 데이터
        F_last = pivot.loc[follower, months[t]]
        F_prev1 = pivot.loc[follower, months[t-1]] if t >= 1 else F_last

        # 핵심 피처 5개만
        features = {
            'f_raw_value': L_raw,                           # lag 시점 선행 값
            'f_leader_last': L_last,                        # 최근 선행 값
            'f_follower_last': F_last,                      # 최근 후행 값
            'f_leader_momentum': (L_last - L_prev1) / (L_prev1 + 1) if t >= 1 else 0,  # 선행 모멘텀
            'f_follower_momentum': (F_last - F_prev1) / (F_prev1 + 1) if t >= 1 else 0  # 후행 모멘텀
        }

        features_list.append(features)

    # 피처 데이터프레임 생성
    features_df = pd.DataFrame(features_list)

    # pairs_meta에 피처 추가
    pairs_meta_with_features = pd.concat([pairs_meta.reset_index(drop=True),
                                          features_df.reset_index(drop=True)],
                                         axis=1)

    return pairs_meta_with_features

# 사용법
pairs_meta_enhanced = add_time_series_features(pairs_meta, pivot)
print(pairs_meta_enhanced.head())
print(f"\n추가된 피처: {pairs_meta_enhanced.shape[1] - pairs_meta.shape[1]}개")
print(f"\n추가된 피처 목록:")
print(['f_raw_value', 'f_leader_last', 'f_follower_last', 'f_leader_momentum', 'f_follower_momentum'])
````

````python
import os
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lightgbm import LGBMRegressor
from tqdm import tqdm

# ============================================================
# pairs_meta_enhanced 기반 학습 데이터 생성 (핵심 5개 피처)
# ============================================================

def build_training_data_enhanced(pivot, pairs_meta_enhanced, months_dt):
    """
    pairs_meta_enhanced를 활용한 학습데이터 생성

    Feature 구성:
      - 핵심 시계열 피처 5개
      - pairs_meta_enhanced에 이미 있는 메타 피처 활용
    """
    months = months_dt
    n_months = len(months)
    rows = []

    for idx, meta_row in tqdm(pairs_meta_enhanced.iterrows(), total=len(pairs_meta_enhanced)):
        leader = meta_row['leading_item_id']
        follower = meta_row['following_item_id']
        lag = int(meta_row['f_lag'])

        # 각 시점에 대해 학습 샘플 생성
        for t in range(lag, n_months - 1):
            ref_idx = t - lag
            target_month = months[t + 1]

            # Leader 데이터
            L_raw = pivot.loc[leader, months[ref_idx]]
            L_last = pivot.loc[leader, months[t]]
            L_prev1 = pivot.loc[leader, months[t-1]] if t >= 1 else L_last

            # Follower 데이터
            F_last = pivot.loc[follower, months[t]]
            F_prev1 = pivot.loc[follower, months[t-1]] if t >= 1 else F_last

            # 피처 계산
            features = {
                "leading_item_id": leader,
                "following_item_id": follower,
                "target_month": target_month,
                "value": pivot.loc[follower, target_month],

                # 핵심 5개 피처
                "f_raw_value": L_raw,
                "f_leader_last": L_last,
                "f_follower_last": F_last,
                "f_leader_momentum": (L_last - L_prev1) / (L_prev1 + 1) if t >= 1 else 0,
                "f_follower_momentum": (F_last - F_prev1) / (F_prev1 + 1) if t >= 1 else 0,
            }

            # pairs_meta_enhanced의 메타 피처 추가
            for col in pairs_meta_enhanced.columns:
                if col not in ['leading_item_id', 'following_item_id',
                              'f_raw_value', 'f_leader_last', 'f_follower_last',
                              'f_leader_momentum', 'f_follower_momentum']:
                    features[col] = meta_row[col]

            rows.append(features)

    return pd.DataFrame(rows)


# ============================================================
# 학습용 피처 컬럼 정의 (핵심 5개만)
# ============================================================
feature_cols = [
    "f_raw_value",
    "f_leader_last",
    "f_follower_last",
    "f_leader_momentum",
    "f_follower_momentum"
]

# ============================================================
# 학습 시작
# ============================================================
print("=== 학습 데이터 생성 시작 ===")
df_train = build_training_data_enhanced(pivot, pairs_meta_enhanced, months_dt)

print(f'생성된 학습 데이터의 shape: {df_train.shape}')

if df_train.empty:
    print("오류: 학습 데이터가 없습니다.")
    submission = pd.DataFrame(columns=['leading_item_id', 'following_item_id', 'value'])
else:
    # 결측치 처리
    df_train[feature_cols] = df_train[feature_cols].fillna(0)

    train_X = df_train[feature_cols].values
    train_y = df_train["value"].values

    # LGBM 모델 학습
    print("\n=== LGBM 모델 학습 시작 ===")
    reg = LGBMRegressor(
        random_state=42,
        n_estimators=500,
        learning_rate=0.05,
        num_leaves=31
    )

    reg.fit(train_X, train_y)
    print("✅ 모델 학습 완료")

    # ============================================================
    # Feature Importance 시각화
    # ============================================================
    importance_gain = reg.booster_.feature_importance(importance_type='gain')

    # Gain 기준 정렬
    idx = np.argsort(importance_gain)

    plt.figure(figsize=(10, 5))
    plt.barh(np.array(feature_cols)[idx], importance_gain[idx])
    plt.title("LightGBM Feature Importance (Gain) - 5 Features")
    plt.xlabel("Gain")
    plt.tight_layout()
    plt.show()

    # Feature Importance 출력
    print("\n=== Feature Importance ===")
    feat_imp_df = pd.DataFrame({
        'feature': feature_cols,
        'gain': importance_gain
    }).sort_values('gain', ascending=False)
    print(feat_imp_df)


# ============================================================
# 예측 (2025년 8월)
# ============================================================
print("\n=== 2025년 8월 예측 시작 ===")

preds = []
t_last = len(months_dt) - 1

for idx, meta_row in tqdm(pairs_meta_enhanced.iterrows(), total=len(pairs_meta_enhanced)):
    leader = meta_row['leading_item_id']
    follower = meta_row['following_item_id']
    lag = int(meta_row['f_lag'])

    ref_idx = t_last - lag

    # Leader 데이터
    L_raw = pivot.loc[leader, months_dt[ref_idx]]
    L_last = pivot.loc[leader, months_dt[t_last]]
    L_prev1 = pivot.loc[leader, months_dt[t_last-1]] if t_last >= 1 else L_last

    # Follower 데이터
    F_last = pivot.loc[follower, months_dt[t_last]]
    F_prev1 = pivot.loc[follower, months_dt[t_last-1]] if t_last >= 1 else F_last

    # 예측용 피처 배열 (5개만)
    X_row = np.array([[
        L_raw,                                    # f_raw_value
        L_last,                                   # f_leader_last
        F_last,                                   # f_follower_last
        (L_last - L_prev1) / (L_prev1 + 1),     # f_leader_momentum
        (F_last - F_prev1) / (F_prev1 + 1)      # f_follower_momentum
    ]])

    # 결측치 처리
    X_row = np.nan_to_num(X_row, nan=0.0)

    # 예측
    pred = reg.predict(X_row)[0]
    pred = int(max(0, round(pred)))

    preds.append({
        "leading_item_id": leader,
        "following_item_id": follower,
        "value": pred
    })

# ============================================================
# 최종 제출 파일 저장
# ============================================================
date_str = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
file_name = f"sy_5features_{date_str}.csv"
submission = pd.DataFrame(preds)
submission.to_csv(file_name, index=False, encoding='utf-8-sig')

print(f"\n✅ 예측 완료! {file_name} 저장됨")
print(f"제출 파일 shape: {submission.shape}")
print(f"\n미리보기:")
print(submission.head(10))

from google.colab import files

# 제출 파일 다운로드
files.download(file_name)
````
